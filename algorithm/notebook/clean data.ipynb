````
string[] lines = data.Split("\n");
string[] keywords = new string[] {"Basic", "Easy", "Medium", "Hard", "Accuracy", "Submission"};

var questionNames = lines.Where(line => !keywords.Any(keyword => line.Contains(keyword))).ToList();

// Remove any empty lines
questionNames = questionNames.Where(name => !string.IsNullOrWhiteSpace(name)).ToList();

// Write the question names to a text file
System.IO.File.WriteAllLines("output.txt", questionNames);
````

````
// Read the file
string[] lines = System.IO.File.ReadAllLines("output.txt");

// Define the types of tree questions to focus on
string[] treeTypes = new string[] {"Binary", "Spiral", "Sum", "Mirror", "Perfect"};

// Count the occurrences of each tree type
Dictionary<string, int> treeTypeCounts = new Dictionary<string, int>();
foreach (var treeType in treeTypes)
{
    treeTypeCounts[treeType] = lines.Count(line => line.Contains(treeType));
}

// Calculate the percentage of occurrences for each tree type
Dictionary<string, double> treeTypePercentages = new Dictionary<string, double>();
foreach (var treeType in treeTypes)
{
    treeTypePercentages[treeType] = (double)treeTypeCounts[treeType] / lines.Length * 100;
}

// Print the results
foreach (var treeType in treeTypes)
{
    Console.WriteLine($"- {treeType}: {treeTypeCounts[treeType]} occurrences ({treeTypePercentages[treeType]:0.00}% of total)");
}
````
